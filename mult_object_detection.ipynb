```python
import cv2 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
img = cv2.imread('threeThings.jpg')
```

```python
plt.imshow(img)
```

```python
#get gray
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
#get rgb
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img_gray)
```

```python
plt.imshow(img_rgb)
```

```python
#detect a cat  

cat_data = cv2.CascadeClassifier('haarcascade_frontalcatface_extended.xml') 
  
cats_found = cat_data.detectMultiScale(img_gray,  
                                   minSize =(20, 20))

cats_amount_found = len(cats_found)

if cats_amount_found != 0:   
    for (x, y, width, height) in cats_found: 
        cv2.rectangle(img_rgb, (x, y),  
                      (x + height, y + width),  
                      (0, 255, 0), 5) 
        cv2.putText(img_rgb, "CAT", (x, y-10), cv2.FONT_HERSHEY_SIMPLEX, 1, (0,0,255), 2)
```

```python
#detect a face
smile_data = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
smiles_found = smile_data.detectMultiScale(img_gray, minSize = (30,30))
smiles_amount_found = len(smiles_found)

if smiles_amount_found != 0:
    for (x,y,w,h) in smiles_found:
        cv2.rectangle(img_rgb, (x,y), 
                        (x +h, y+w), 
                        (0,255,0), 5)
        cv2.putText(img_rgb, "FACE", (x, y-10), cv2.FONT_HERSHEY_SIMPLEX, 1, (0,0,255), 2)
```

```python
#detect stop sign
sign_data = cv2.CascadeClassifier('stop_data.xml')
signs_found = sign_data.detectMultiScale(img_gray, minSize = (30,30))
signs_amount_found = len(signs_found)

if signs_amount_found !=0:
    for (x,y,w,h) in signs_found:
        cv2.rectangle(img_rgb, (x,y), 
                        (x +h, y+w), 
                        (0,255,0), 5)
    cv2.putText(img_rgb, "Stop Sign", (x, y-10), cv2.FONT_HERSHEY_SIMPLEX, 1, (0,0,255), 2)
```

```python
plt.subplot(1, 1, 1) 
plt.imshow(img_rgb) 
plt.show() 
```
